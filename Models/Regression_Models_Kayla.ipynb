```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import numpy as np  # Importing NumPy library
import pandas as pd  # Importing Pandas library
import matplotlib.pyplot as plt  # Importing Matplotlib library's "pyplot" module
import seaborn as sns  # Imorting Seaborn library
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#load and preprocess the data
#need to label my X_test, X_train, Y_test, Y_train
y_train = pd.read_csv(r'C:\Users\14037\Downloads\Biomed 415\Kaylas_Models\BMEN-415\Data\_split datasets\regression_ytrain.csv')
X_train = pd.read_csv(r'C:\Users\14037\Downloads\Biomed 415\Kaylas_Models\BMEN-415\Data\_split datasets\regression_Xtrain.csv')
y_test = pd.read_csv(r'C:\Users\14037\Downloads\Biomed 415\Kaylas_Models\BMEN-415\Data\_split datasets\regression_ytest.csv')
X_test = pd.read_csv(r'C:\Users\14037\Downloads\Biomed 415\Kaylas_Models\BMEN-415\Data\_split datasets\regression_Xtest.csv')

```

```python
#Ridge Regression Model


rr = Ridge(alpha=10)
rr.fit(X_train, y_train)

y_train_pred1 = rr.predict(X_train)
y_test_pred1 = rr.predict(X_test)

#r1 = r1_score(y_test, y_test_pred1)

print('Training MSE:', mean_squared_error(y_train, y_train_pred1))
print('Testing MSE:', mean_squared_error(y_test, y_test_pred1))

fig = plt.figure()
ax1 = fig.add_subplot(111)

plt.scatter(y_test, y_test_pred1)
plt.title('Ridge Regression Model\nPredicted vs. Actual (test data)')
plt.xlabel('Actual')
plt.ylabel('Predicted')

plt.show()
```

```python
# XGBoost model
xgb = XGBRegressor()
xgb.fit(X_train, y_train)

y_train_pred2 = xgb.predict(X_train)
y_test_pred2 = xgb.predict(X_test)

print('Training MSE:', mean_squared_error(y_train, y_train_pred2))
print('Testing MSE:', mean_squared_error(y_test, y_test_pred2))

fig = plt.figure()
ax1 = fig.add_subplot(111)

plt.scatter(y_test, y_test_pred2)
plt.title('XGBoost model\nPredicted vs. Actual (test data)')
plt.xlabel('Actual')
plt.ylabel('Predicted')

plt.show()
```

```python
# Gradient Boosting Regression model
gbr = GradientBoostingRegressor(n_estimators=50, learning_rate=0.01, random_state=24)
gbr.fit(X_train, y_train)

y_train_pred3 = gbr.predict(X_train)
y_test_pred3 = gbr.predict(X_test)

print('Training MSE:', mean_squared_error(y_train, y_train_pred3))
print('Testing MSE:', mean_squared_error(y_test, y_test_pred3))

fig = plt.figure()
ax1 = fig.add_subplot(111)

plt.scatter(y_test, y_test_pred3)
plt.title('Gradient Boosting Regression model\nPredicted vs. Actual (test data)')
plt.xlabel('Actual')
plt.ylabel('Predicted')

plt.show()
```
