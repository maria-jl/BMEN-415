```python
#import headers

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
#import data

data = pd.read_csv("wdbc.csv")

x_train = pd.read_csv("classification_Xtrain.csv")
x_test  = pd.read_csv("classification_Xtest.csv")
y_train = pd.read_csv("classification_ytrain.csv")
y_test  = pd.read_csv("classification_ytest.csv")
```

KNN

```python
#finding best k values

K = []
training = []
test = []
scores = {}
for k in range(2, 40):
    clf = KNeighborsClassifier(n_neighbors = k)
    clf.fit(x_train, y_train)

    training_score = clf.score(x_train, y_train)
    test_score = clf.score (x_test, y_test)
    K.append(k)

    training.append(training_score)
    test.append(test_score)
    scores[k] = [training_score, test_score]

for keys, values in scores.items():
    print(keys, ':', values)
    
```

```python
#plotting k values vs training data and testing data

plt.scatter(K, training, color = 'k')
plt.scatter(K, test, color = 'g')
plt.show()
```

```python
#knn model

clf = KNeighborsClassifier(n_neighbors = 9)
clf.fit(x_train, y_train)
training_score = clf.score(x_train, y_train)
test_score = clf.score(x_test, y_test)
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
print("Training Score: ", training_score)
print("Test Score", test_score)
confusion = confusion_matrix(y_test, y_test_pred)
sns.heatmap(confusion, square = True, annot = True, cbar = False, cmap = 'icefire', 
            xticklabels = ['benign', 'malignant'], yticklabels = ['benign', 'malignant'])
plt.xlabel('predicted value')
plt.ylabel('true value')
```

ExtraTrees

```python
#finding best number of trees

T = []
training = []
test = []
scores = {}
for t in range(10, 140):
    clf = ExtraTreesClassifier(n_estimators = t, random_state = None)
    clf.fit(x_train, y_train)

    training_score = clf.score(x_train, y_train)
    test_score = clf.score (x_test, y_test)
    T.append(t)

    training.append(training_score)
    test.append(test_score)
    scores[t] = [training_score, test_score]

for keys, values in scores.items():
    print(keys, ':', values)
    
```

```python
#extratrees model

clf = ExtraTreesClassifier(n_estimators = 117, random_state = None)
clf.fit(x_train, y_train)
training_score = clf.score(x_train, y_train)
test_score = clf.score(x_test, y_test)
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
print("Training Score: ", training_score)
print("Test Score", test_score)
confusion = confusion_matrix(y_test, y_test_pred)
sns.heatmap(confusion, square = True, annot = True, cbar = False, cmap = 'icefire', 
            xticklabels = ['benign', 'malignant'], yticklabels = ['benign', 'malignant'])
plt.xlabel('predicted value')
plt.ylabel('true value')
```
